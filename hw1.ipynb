```python
import numpy as np
import matplotlib.pyplot as plt
import cmocean
```

```python
L = 10e3
H = 100
f = 1e-4
x = np.linspace(-4*L,4*L,1000)
z = np.linspace(0,-H,1000)
x,z = np.meshgrid(x,z)
drho = 1
rho_0 = 1000
g = 9.81
np.tanh(-1e7)
```

```python
V = -g*drho/(rho_0*f*L)*1/(np.cosh(x/L)**2)*(z+H)
rho = rho_0+drho*(np.tanh(x/L)-z/H)
plt.figure(figsize = (12,8))
plt.pcolormesh(x/1e3,z,V, cmap = cmocean.cm.balance, vmin = -1, vmax = 1)
plt.colorbar(label = 'V [m/s]')
plt.contour(x/1e3,z,rho,levels=np.linspace(np.min(rho),np.max(rho),12),colors='black',alpha=0.5)
plt.xlabel('x [km]')
plt.ylabel('z [m]')
# plt.hlines(-100,-40,40,linestyle='--')
# plt.annotate(-40,100,'ML depth')
plt.savefig('hw1b.png')
```

```python
x_surf = np.linspace(-4*L,4*L)
eta = -(drho*H/rho_0)*(1+np.tanh(x_surf/L))
plt.figure(figsize = (12,8))
plt.plot(x_surf/1e3,eta)
plt.xlabel('x [km]')
plt.ylabel('SSH anomaly [m]')
plt.savefig('hw1c.png')
```

```python
L = 10e3
H = 100
f = 1e-4
x = np.linspace(-4*L,4*L,1000)
z = np.linspace(0,-H,1000)
x,z = np.meshgrid(x,z)
V = -g*drho/(rho_0*f*L)*1/(np.cosh(x/L)**2)*(z+H)
rho = rho_0+drho*(np.tanh(x/L)-z/H)
chi_x = (g*drho**2/(rho_0**2*f*L**2))*((1/np.cosh(x/L)**4)/(1+(drho/rho_0)*(np.tanh(x/L)-z/H)))
chi_z = -(2*g*drho**2/(rho_0**2*f*H*L**2)*(z+H)*(np.tanh(x/L)*(1/np.cosh(x/L)**2))/(1+(drho/rho_0)*(np.tanh(x/L)-z/H)))
chi_background = -f*drho/(H*rho)
fig, axs = plt.subplots(3,1,figsize = (12,12), constrained_layout=True)
im1 = axs[0].pcolormesh(x/1e3,z,chi_x,cmap=cmocean.cm.amp)
im2 = axs[1].pcolormesh(x/1e3,z,chi_z,cmap=cmocean.cm.balance)
im3 = axs[2].pcolormesh(x/1e3,z,chi_background,cmap=cmocean.cm.amp)
cbar1 = fig.colorbar(im1,ax = axs[0], location='bottom', shrink = 0.5)
cbar1.set_label('PV')
# cbar1.ax.tick_params(labelsize=16)
cbar2 = fig.colorbar(im2,ax = axs[1], location='bottom', shrink = 0.5)
cbar2.set_label('PV')
# cbar2.ax.tick_params(labelsize=16)
cbar3 = fig.colorbar(im3,ax = axs[2], location='bottom', shrink = 0.5)
cbar3.set_label('PV')
# cbar3.ax.tick_params(labelsize=16)
axs[0].set_title('Horizontal vorticity component of Ertel PV')
axs[1].set_title('Vetical vorticity component of Ertel PV')
axs[2].set_title('Background component of Ertel PV')
for ax in axs:
    ax.set_xlabel('x [km]')
    ax.set_ylabel('y [m]')
    
# plt.colorbar()axs[0].set_title('Horizontal velocity component of Ertel PV')
plt.savefig('hw1f.png')
```
