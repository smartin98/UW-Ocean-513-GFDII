```python
import numpy as np
import matplotlib.pyplot as plt
import cmocean
```

```python
f0 = -1e-4
omega = 2*np.pi/86164.1
L = 1000e3
theta = np.arcsin(f0/(2*omega)) 
R = 6378e3 # radius of Earth
beta = 2*omega*np.cos(theta)/R
f0/beta/1e3
```

```python
N_contours = 21
x0 = L*np.zeros(N_contours)
y0 = np.linspace(-1000e3,1000e3,N_contours)
x = np.linspace(-5*L,5*L,1000)
y = np.zeros((N_contours,1000))
for i in range(N_contours):
    y[i,:] = -f0/beta+(f0/beta+y0[i])*np.exp((x0[i]**2-x**2)/L**2)
print(np.max(np.exp((x0[i]**2-x**2)/L**2))/1e3)
```

```python
for i in range(N_contours):
    plt.plot(x/1e3,y[i,:]/1e3-6e3,color='black')
plt.plot(x/1e3,y[10,:]/1e3-6e3,color='red')
plt.xlabel('x [km]')
plt.ylabel('y [km]')
plt.savefig('acc.png')
```
